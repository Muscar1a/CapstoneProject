```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import os 
%matplotlib inline
```

```python
file = "healthcare-dataset-stroke-data.csv"
```

```python
df = pd.read_csv(file)
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['bmi'].replace(to_replace=np.nan, value=df['bmi'].mean(), inplace=True)
```

```python
df['bmi'].isna().sum()
```

Replace missing value in "bmi" by its mean value.

```python
df.describe()
```

```python
df['gender'].value_counts()
sns.set_theme(style='darkgrid')
ax = sns.countplot(data=df,x='gender')
```

```python
print(df['ever_married'].value_counts())
sns.set_theme(style="darkgrid")
ax = sns.countplot(data=df, x='ever_married')
```

```python
print(df['work_type'].value_counts())
sns.set_theme(style='darkgrid')
ax = sns.countplot(data=df, x='work_type')
```

```python
print(df['Residence_type'].value_counts())
sns.set_theme(style='darkgrid')
ax = sns.countplot(data=df, x='Residence_type')
```

```python
print(df['smoking_status'].value_counts())
sns.set_theme(style='darkgrid')
ax = sns.countplot(data=df, x='smoking_status')
```

```python
print(df['stroke'].value_counts())
sns.set_theme(style='darkgrid')
ax = sns.countplot(data=df, x='stroke')
```

We can see that our data is imbalance

```python
# fig = plt.figure(figsize=(7, 7))
sns.displot(df['avg_glucose_level'], color='green', label='avg_glucose_level', kde=True)
plt.legend()
```

```python
sns.distplot(df['bmi'], color='orange', label='bmi', kde=True)
plt.legend()
```

```python
plt.figure(figsize=(7,7))
sns.distplot(df[df['stroke'] == 0]['bmi'],color='green', kde=True)
sns.distplot(df[df['stroke'] == 1]['bmi'],color='red', kde=True)

plt.title('No Stroke vs Stroke by BMI',fontsize=15)
plt.xlim([10,100])
```

```python
plt.figure(figsize=(12,10))
sns.distplot(df[df['stroke']==0]['avg_glucose_level'], color='green', kde=True)
sns.distplot(df[df['stroke']==1]['avg_glucose_level'], color='red', kde=True)

plt.title("No Stroke vs Stroke by Avg Glucose Level")
plt.xlim([30, 330])
```

```python
plt.figure(figsize=(12,10))
sns.distplot(df[df['stroke']==0]['age'], color='green')
sns.distplot(df[df['stroke']==1]['age'], color='red')

plt.title("No Stroke vs Stroke by Age")
plt.xlim([18,100])
```

```python
fig = plt.figure(figsize=(7,7))
graph = sns.scatterplot(data=df, x='age', y='bmi', hue='gender')
graph.axhline(y=25, linewidth=4, color='r', linestyle='--')
```

```python
fig = plt.figure(figsize=(7,7))
graph = sns.scatterplot(data=df, x='age', y='avg_glucose_level', hue='gender')
graph.axhline(y=150, linewidth=4, color='r', linestyle='--')
```
