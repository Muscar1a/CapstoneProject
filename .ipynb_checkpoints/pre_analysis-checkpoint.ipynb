# Read data

```python
import numpy as np  # type: ignore
import pandas as pd # type: ignore
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("creditcard.csv")
```

```python
data.info()
```

```python
data.describe()
```

From **Time** column, we know that there are **284807** transactions during **172792** seconds.

```python
total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum() / data.isnull().count() * 100).sort_values(ascending=False)
pd.concat([percent, total], axis=1, keys = ["percent", "total"])
```

There is no missing data in our dataset

## Checking the balance in dataset

```python
temp = data['Class'].value_counts()
df = pd.DataFrame({"Class":temp.index, "Values":temp.values})

index = [i for i in temp.index]
values = [i for i in temp.values]
plt.bar(temp.index, temp.values, color='maroon')
plt.xticks(index)
plt.xlabel("Class")
plt.ylabel("Number of transactions")
plt.title("Credit card fraud class (Not Fraud = 0, Fraud = 1)")
plt.show()

plt.pie(temp.values, labels=["Not Fraud", "Fraud"], autopct="%1.1f%%", colors=['maroon', 'yellow'])
plt.legend(['Not Fraud', 'Fraud'])
plt.title("Credit card fraud class")
plt.show()
```

We only have **0.2%** of fraud transactions while **99.8%** of not fraud transactions, so we can claim that our dataset is unbalance.

=> The dataset is unbalance.

Since our dataset is unbalance, we can't use metric like accuracy_score to evaluate our prediction models.

# Data exploration

#### Feature Distribution 

```python
class_0 = data.loc[data['Class'] == 0]['Time']
class_1 = data.loc[data['Class'] == 1]['Time']
plt.figure(figsize=(10, 6))

sns.kdeplot(class_0, label='Not Fraud', fill=True)
sns.kdeplot(class_1, label='Fraud', fill=True)

plt.title("Credit Card Transactions Time Density Plot")
plt.xlabel("Time(s)")
plt.xlim(0, 172792)
plt.yticks([])
plt.legend()

plt.show()
```

In some periods of time, the distribution of fraudulent transactions is higher than valid transactions, including low real transaction times (night in the place where our dataset belong to).

Let's look into details to the time distribution of both class transaction and aggregated values of transactions count and amount per hour.
Assume that the initial time unit is second (based on our observation and inference).

```python
data['Hour'] = data['Time'].apply(lambda x:np.floor(x / 3600))

tmp = data.groupby(['Hour', 'Class'])['Amount'].aggregate(['min', 'max', 'count', 'sum', 'mean', 'median', 'var']).reset_index()

df = pd.DataFrame(tmp)
df.columns = ['Hour', 'Class', 'Min', 'Max', 'Transactions', 'Sum', 'Mean', 'Median', 'Var']
df.tail(10)
```

* Note that green line represents for class 0, which is non fraud, while red is for fraudulent transactions.

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18,6))
s = sns.lineplot(ax=ax1, x='Hour', y='Sum', data=df.loc[df['Class']==0], color='green')
s = sns.lineplot(ax=ax2, x='Hour', y='Sum', data=df.loc[df['Class']==1], color='red')
plt.suptitle("Total Amount")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18,6))
s = sns.lineplot(ax=ax1, x='Hour', y='Transactions', data=df.loc[df['Class']==0], color='green')
s = sns.lineplot(ax=ax2, x='Hour', y='Transactions', data=df.loc[df['Class']==1], color='red')
plt.suptitle("Total number of Transactions")
plt.show()
```

### Features correlation

```python
plt.figure(figsize=(14, 14))
plt.title('Creditcard Transactions features correlation plot (Pearson)')
corr = data.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, linewidths=.1, cmap="Reds")
plt.show()
```

There is no notable correlation between **V1-V28**. There are some noticeable correlations between **Time**(inverse correlation with **V3**) and **Amount**(correlation with **V7**, **V20**, inverse correlation with **V2**, **V5**).

Let's plot some correlated and inverse correlated values on same graph to see the trend.

```python
s = sns.lmplot(x='V20', y='Amount', data=data, hue='Class', fit_reg=True, scatter_kws={'s':2})
s = sns.lmplot(x='V7', y='Amount', data=data, hue='Class', fit_reg=True, scatter_kws={'s':2})
plt.show()
```

As we can see, the two couples of features show positive slope. It's mean that they are correlated. The regression line for **Class 0** have positive slope, while the one for **Class 1** is much smaller positive slope.

```python
s = sns.lmplot(x='V2', y='Amount', data=data, hue='Class', fit_reg=True, scatter_kws={'s':2})
s = sns.lmplot(x='V5', y='Amount', data=data, hue='Class', fit_reg=True, scatter_kws={'s':2})
plt.show()
```

Similar to above, two couples of features show a downward in the correlation between **Amount** and **V2** or **V5**. While **Class 0** represent a regression line with steeper slope, **Class 1** show very small negative slope.

### Features density plot

* Blue line represents non fraudulent transactions.
* Orange line for fraudulent transactions.

```python
var = data.columns.values

i = 0
t0 = data.loc[data['Class'] == 0]
t1 = data.loc[data['Class'] == 1]

sns.set_style('whitegrid')
plt.figure()
fix, ax = plt.subplots(8, 4, figsize=(16, 28))

for feature in var:
    i += 1
    plt.subplot(8, 4, i)
    sns.kdeplot(t0[feature], bw_method=0.5, label='Class 0', color='blue')
    sns.kdeplot(t1[feature], bw_method=0.5, label='Class 1', color='orange')
    plt.xlabel(feature, fontsize=12)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
    
```

For some features, we observe a good selectivity of distribution for 2 values of **Class**: 
* **V4**, **V9**, **V10**, **V11** have clearly separated distributions.
* **V12**, **V14**, **16**,**17**, **V18** are partially separated.
* **V1, V2, V3, V10** have quite disctinct profile.
* **V15, V24, V26, V28** have similar profiles for two values of Class.

In general, the feature distribution for legitimate transactions (**Class 0**) us centered around 0. 
At the same time, fraudulent transactions (**Class 1**) have skewed distribution.
